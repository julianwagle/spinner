```python
import sys
#print(sys.getrecursionlimit())
sys.setrecursionlimit(15000)
import random
import matplotlib.pyplot as plt

class Going_Gambelling():

    # TODO: Add in multi-bet feature to track both high-low and red-black

    # ADJEST # BELOW # VALUES # FREQUENTLY # FOR # EXPIRAMENTATION # 
    max_bet_value = float(5000.00)
    wallet_starting_value = float(10000.00)
    wallet_minimum_authorized_value = float(10.00)
    desired_return_per_bet = float(10)
    max_daily_spins = float(400)
    table_losses_before_entry = 7
    total_days_of_play = float(127)
    # ADJEST # ABOVE # VALUES # FREQUENTLY # FOR # EXPIRAMENTATION # 

    wallet_total_value = wallet_starting_value
    sanity_checker_wins = []
    sanity_checker_losses = []
    sanity_checker_total = []
    sanity_checker_deviations = []
    previous_bet_sizes = []
    total_daily_spins = 0
    table_spin_loss_counter = 0
    total_days_played = 0

    table_slots_all = list(range( 1 , 39 ))
    table_slots_high = list(range( 1 , 19 )) # 1 to 18
    table_slots_low = list(range( 19 , 37 )) # 19 to 37
    table_slots_zero = list(range( 37 , 39 )) # 37 to 38
    table_slots_red = [1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]
    table_slots_black = [2, 4, 6, 8, 10, 11, 13, 15, 17, 20, 22, 24, 26, 28, 29, 31, 33, 35]
    table_slots_green = [37, 38]


    def start_gambelling(self):
        #print("at the casino now.")
        return self.sit_at_the_table()


    def sit_at_the_table(self): 
        #print("sitting at the table.")
        broke = bool(self.wallet_total_value <= self.wallet_minimum_authorized_value)
        while not broke:

            table_spin_results = self.watch_the_table_spin()
            
            if isinstance(table_spin_results, float):
                table_spin_loss = bool(table_spin_results > 18)
                table_spin_win = bool(table_spin_results  <= 18)
                self.sanity_checker_total.append(1)

                if table_spin_loss: 
                    self.table_spin_loss_counter += 1
                    self.sanity_checker_losses.append(1)
                elif table_spin_win: 
                    self.table_spin_loss_counter = 0
                    self.sanity_checker_wins.append(1)
                else: assert 1 < 0
                
                time_to_bet = bool(self.table_losses_before_entry == self.table_spin_loss_counter)
                if time_to_bet:
                    #print(f"Its time to bet on spin number {self.total_daily_spins}")
                    self.table_spin_loss_counter = float(0)
                    self.place_a_bet()


    def watch_the_table_spin(self):
        self.total_daily_spins += 1
        my_day_is_over = bool(self.total_daily_spins > self.max_daily_spins)
        broke = bool(self.wallet_total_value <= self.wallet_minimum_authorized_value)
        #print(f"watcing the table spin for time number {self.total_daily_spins}")
        if broke or my_day_is_over:
            self.head_home()
        else:
            table_spin_results = random.choice(self.table_slots_all)
            return float(table_spin_results)


    def verify_sanity(self):
        #print("verifying sanity ...")
        total_spins = float(sum(self.sanity_checker_total))
        total_wins = float(sum(self.sanity_checker_wins))
        total_losses = float(sum(self.sanity_checker_losses)) 
        assert total_spins == total_wins + total_losses
        winning_frequency = float(total_wins / total_spins) * 100
        estimated_likelyhood_of_winning = float(len(self.table_slots_high)/len(self.table_slots_all)) * 100
        deviation_from_standard = estimated_likelyhood_of_winning - winning_frequency
        deviation_from_standard_polished = round(deviation_from_standard,3)
        self.sanity_checker_deviations.append(deviation_from_standard)
        deviation_average = float(sum(self.sanity_checker_deviations) / len(self.sanity_checker_deviations))
        deviation_average_polished = round(deviation_average,3)
        sanity_check_message = str(f"CURRENT DEVIATION:{deviation_from_standard_polished} AVERAGE DEVIATION:{deviation_average_polished}")
        #print(f"i'm only {deviation_average_polished}% insane so far")
        return sanity_check_message


    def determine_bet_size(self):
        #print("determining bet size.")
        sum_of_losses = int(sum(self.previous_bet_sizes))
        desired_back_payment = len(self.previous_bet_sizes) * self.desired_return_per_bet
        bet_size = self.desired_return_per_bet + desired_back_payment + sum_of_losses
        reached_table_max = bool(bet_size > self.max_bet_value)
        if reached_table_max: 
            print("I have reached the table max.")
            bet_size = self.previous_bet_sizes[-1]
            #self.take_a_fat_l()
            return bet_size 
        else:
            return bet_size 


    def place_a_bet(self):
        self.verify_sanity()
        #print("its betting time!")

        bet_size = self.determine_bet_size()
        if isinstance(bet_size, float):
            
            table_spin_results = self.watch_the_table_spin()    
            if isinstance(table_spin_results, float):
                table_spin_loss = bool(table_spin_results > 18)
                table_spin_win = bool(table_spin_results  <= 18)
                self.sanity_checker_total.append(1)

                if table_spin_loss: 
                    self.sanity_checker_losses.append(1)                    
                    self.wallet_total_value -= bet_size
                    self.previous_bet_sizes.append(bet_size)
                    #print(f"I lost --${bet_size}--")
                    return self.place_a_bet()

                elif table_spin_win: 
                    self.sanity_checker_wins.append(1)                    
                    self.wallet_total_value += bet_size
                    self.previous_bet_sizes = []
                    # print(f"I won ++${bet_size}++")
                    return self.sit_at_the_table()
                else: assert 1 < 0
                


    def take_a_fat_l(self):
        print("taking a fat l.")
        self.table_spin_loss_counter = 0
        self.previous_bet_sizes = []
        return self.sit_at_the_table()


    def head_home(self):
        self.previous_bet_sizes = []
        self.total_daily_spins = 0
        self.table_spin_loss_counter = 0
        self.total_days_played += 1
        broke = bool(self.wallet_total_value <= self.wallet_minimum_authorized_value)
        if broke:
            print(f"You went broke on day {self.total_days_played}")
            return sys.exit()
        else:
            the_time_frame_is_over = bool(self.total_days_played == self.total_days_of_play)
        if the_time_frame_is_over:
            self.perform_analysis()
        else:
            self.start_gambelling()


    def perform_analysis(self):
        total_gains = self.wallet_total_value - self.wallet_starting_value
        average_daily_return = total_gains / self.total_days_of_play
        print(self.verify_sanity())
        print("##############################################################################################################")
        print(f"The time period is over. Your wallet is at ${self.wallet_total_value}. You received an average daily return of ${average_daily_return}")
        print("##############################################################################################################")   
        self.sanity_checker_wins = []
        self.sanity_checker_losses = []
        self.sanity_checker_total = []
        self.sanity_checker_deviations = []
        self.wallet_total_value = float(10000.00)
        return sys.exit()

Go_Gambelling = Going_Gambelling()
Go_Gambelling.start_gambelling()



```
